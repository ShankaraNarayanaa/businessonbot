```python
import pandas as pd
import collections
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
data = pd.read_csv('tweets.csv')
train = data.loc[:1000]
test = data.loc[1000:]
```

```python
import re

def preprocess_text(sentence):
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','', sentence['OriginalTweet'])
    text = re.sub('@[^\s]+','', text)
    text = re.sub('&[^\s]+;', '', text)
    text = re.sub('[^a-zA-Zа-яА-Я1-9]+', ' ', text)
    text = re.sub(' +',' ', text)
    return text.strip()

preprocess = train
preprocess['Cleaned'] = preprocess.apply(preprocess_text, axis=1)

preprocess.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
def print_sentiment_scores(ugc):
    snt = analyzer.polarity_scores(ugc['Cleaned'])
    return snt['compound']
```

```python
compound = train
compound['Predicted']=compound.apply(print_sentiment_scores, axis=1)

compound.head()
```
